Ich lösche die Spalten mit den meisten missing values, für den Rest probiere ich die Data Cleaning Methode (fügt den häufigsten Wert bei leeren Werten ein):

```python
from sklearn.impute import SimpleImputer

categorical_cols = x_train.columns[x_train.isnull().any()]
# For categorical features - use most frequent value 
imputer_cat = SimpleImputer(strategy='most_frequent') 
x_train[categorical_cols] = imputer_cat.fit_transform(x_train[categorical_cols])
df_test[categorical_cols] = imputer_cat.fit_transform(df_test[categorical_cols])
```

```python
x_train.select_dtypes(include='number')
```

manche spalten muss ich von numerisch in kategorisch umwandeln -> numerisch skalieren und kategorisch one-hot-encoding

```python
x_train['MSSubClass'] = x_train['MSSubClass'].astype('object')
df_test['MSSubClass'] = df_test['MSSubClass'].astype('object')
```

```python
x_train.select_dtypes(include='number')
```

```python
column_to_number = ['MasVnrArea', 'GarageYrBlt']
x_train[column_to_number] = x_train[column_to_number].astype(int)
df_test[column_to_number] = df_test[column_to_number].astype(int)
```

```python
# ORDINAL ENCODING (for ordinal variables) 
# Define ordinal mappings 
ordinal_maps = {
    'LotShape': {'IR3': 1, 'IR2': 2, 'IR1': 3, 'Reg': 4},
    'LandSlope': {'Sev': 1, 'Mod': 2, 'Gtl': 3},
    'ExterQual': {'Po': 1, 'Fa': 2, 'TA': 3, 'Gd': 4, 'Ex': 5},
    'ExterCond': {'Po': 1, 'Fa': 2, 'TA': 3, 'Gd': 4, 'Ex': 5},
    'BsmtQual': {'Missing': 0, 'Po': 1, 'Fa': 2, 'TA': 3, 'Gd': 4, 'Ex': 5},
    'BsmtCond': {'Missing': 0, 'Po': 1, 'Fa': 2, 'TA': 3, 'Gd': 4, 'Ex': 5},
    'BsmtFinType1': {'Missing': 0, 'Unf': 1, 'LwQ': 2, 'Rec': 3, 'BLQ': 4, 'ALQ': 5, 'GLQ': 6},
    'BsmtFinType2': {'Missing': 0, 'Unf': 1, 'LwQ': 2, 'Rec': 3, 'BLQ': 4, 'ALQ': 5, 'GLQ': 6},
    'HeatingQC': {'Po': 1, 'Fa': 2, 'TA': 3, 'Gd': 4, 'Ex': 5},
    'KitchenQual': {'Po': 1, 'Fa': 2, 'TA': 3, 'Gd': 4, 'Ex': 5},
    'Functional': {'Sal': 1, 'Sev': 2, 'Maj2': 3, 'Maj1': 4, 'Mod': 5, 'Min2': 6, 'Min1': 7, 'Typ': 8},
    'GarageFinish': {'Missing': 0, 'Unf': 1, 'RFn': 2, 'Fin': 3}
}

for col, mapping in ordinal_maps.items():
    x_train[col] = x_train[col].fillna('Missing')
    x_train[col] = x_train[col].map(mapping)
    df_test[col] = df_test[col].fillna('Missing')
    df_test[col] = df_test[col].map(mapping)
```

```python
x_train.select_dtypes(include='number')
```

```python
from sklearn.preprocessing import StandardScaler
# StandardScaler (Z-score normalization)
scaler_standard = StandardScaler()
cols_to_scale = x_train.select_dtypes(include='number').columns
X_train_standard = scaler_standard.fit_transform(x_train[cols_to_scale])
X_test_scaled = pd.DataFrame(
    scaler_standard.transform(df_test[cols_to_scale]),  # ⚠️ transform, nicht fit_transform!
    columns=cols_to_scale,
    index=df_test.index
)
```

Übrig geblieben sind jetzt die nominal-skalierten Spalten, diese werden folgend One-Hot-encoded

```python
# 1. ONE-HOT ENCODING (for nominal variables) 
# Select nominal categorical variables 
nominal_cols = [
    'MSSubClass', 'MSZoning', 'LandContour', 'LotConfig',
    'Neighborhood', 'Condition1', 'BldgType', 'HouseStyle',
    'RoofStyle', 'Exterior1st', 'Exterior2nd',
    'Foundation', 'BsmtExposure',
    'CentralAir', 'Electrical',
    'GarageType', 'PavedDrive',
    'SaleType', 'SaleCondition'
]
X_train_dummies = pd.get_dummies(x_train[nominal_cols], drop_first=True)
```

```python
X_test_dummies = pd.get_dummies(df_test[nominal_cols], drop_first=True)

# Fehlende Spalten (die im Training vorkamen, im Test aber fehlen) hinzufügen:
missing_cols = set(X_train_dummies.columns) - set(X_test_dummies.columns)
for col in missing_cols:
    X_test_dummies[col] = 0  # einfach mit Nullen auffüllen

# Gleiche Reihenfolge wie im Training
X_test_dummies = X_test_dummies[X_train_dummies.columns]
```

# Feature Selection

```python
from sklearn.feature_selection import VarianceThreshold

# 1. Threshold festlegen
threshold = 0.01

# 2. VarianceThreshold-Objekt erstellen
selector = VarianceThreshold(threshold=threshold)

# 3. Anwenden
X_var_reduced = selector.fit_transform(X_final)
X_test_var_reduced = selector.transform(X_test_final)

# 4. Spaltennamen zurückholen
selected_mask = selector.get_support()
selected_columns = X_final.columns[selected_mask]

# 5. Neuen DataFrame mit ausgewählten Spalten erzeugen
X_selected = pd.DataFrame(X_var_reduced, columns=X_final.columns[selector.get_support()], index=X_final.index)
X_test_selected = pd.DataFrame(X_test_var_reduced, columns=X_test_final.columns[selector.get_support()], index=X_test_final.index)
```

```python
X_selected.shape
```

```python
from sklearn.linear_model import LassoCV

# Standardisieren ist wichtig für Lasso
# Lasso mit automatischem Alpha über Cross-Validation
lasso = LassoCV(cv=5, random_state=42).fit(X_final, y_train)

# Welche Features wurden behalten?
coef_mask = lasso.coef_ != 0
selected_columns = X_final.columns[coef_mask]
X_lasso_selected = X_final[selected_columns]

print(f"Lasso hat {len(selected_columns)} Features behalten")
```

```python
X_lasso_selected
```

```python
from sklearn.linear_model import LinearRegression

# 1. Testdaten auf Lasso-Features reduzieren
X_test_lasso = X_test_final[X_lasso_selected.columns]

imputer = SimpleImputer(strategy='mean')  # oder 'median', 'most_frequent', etc.
X_test_lasso_imputed = pd.DataFrame(
    imputer.fit_transform(X_test_lasso),
    columns=X_test_lasso.columns,
    index=X_test_lasso.index
)

# 2. Trainieren
linreg = LinearRegression()
linreg.fit(X_lasso_selected, y_train)

# 3. Vorhersagen
y_pred = linreg.predict(X_test_lasso_imputed)
```

```python
y_pred
```

```python
# Sicherstellen, dass die Id-Spalte im Original-Testdatensatz noch vorhanden ist
submission = pd.DataFrame({
    'Id': df_test['Id'],           # Originale Id-Spalte aus dem ursprünglichen Testset
    'SalePrice': y_pred         # Vorhersagen aus dem Modell
})
```

## Evaluating results

```python
print("Intercept:", linreg.intercept_)

coeff_df = pd.DataFrame({"Feature": X_lasso_selected.columns, "Coefficient": linreg.coef_})
# Sort dataframe by coefficients.
coeff_df_sorted = coeff_df.sort_values(by="Coefficient", ascending=False)
coeff_df_sorted
```

```python
# Nur Top 20 nach absoluter Koeffizientenhöhe
top_n = 30
top_coeff_df = coeff_df.reindex(coeff_df["Coefficient"].abs().sort_values(ascending=False).index).head(top_n)

# Plot
plt.figure(figsize=(8, 6))
plt.barh(top_coeff_df["Feature"], top_coeff_df["Coefficient"], color="blue")
plt.xlabel("Coefficient Value")
plt.ylabel("Feature")
plt.title(f"Top {top_n} Most Influential Features (Linear Regression)")
plt.gca().invert_yaxis()  # Wichtig: größte oben
plt.tight_layout()
plt.show()
```
